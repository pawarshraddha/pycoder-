### Handling Outliers

```python
sns.boxplot(data=df)
```

Sales and Profit having more outliers
so we can remove it

```python
sns.boxplot(x='Sales' ,data=df)
```

This outlier are remove by using IQR tec

```python
# IQR
Q1 = np.percentile(df['Sales'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(df['Sales'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1


print("Old Shape: ", df.shape)
  
# Upper bound
upper = np.where(df['Sales'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(df['Sales'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
  
print("New Shape: ", df.shape)
  
sns.boxplot(x='Sales', data=df)
```

```python
sns.boxplot(x='Profit' ,data=df)
```

```python
# IQR
Q1 = np.percentile(df['Profit'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(df['Profit'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1


print("Old Shape: ", df.shape)
  
# Upper bound
upper = np.where(df['Profit'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(df['Profit'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
  
print("New Shape: ", df.shape)
  
sns.boxplot(x='Profit', data=df)
```
